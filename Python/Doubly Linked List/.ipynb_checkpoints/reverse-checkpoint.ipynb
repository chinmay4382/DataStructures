```python
class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None
        self.prev = None
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
    def push(self,data):
        new_node=Node(data)
        if(self.head is None):
            self.head=new_node
            new_node.next=None
            new_node.prev=None
            return
        else:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
            new_node.prev = None
            
    def reverse(self):
        curr=self.head
        while(curr.next is not None):
            print(curr.data)
            temp=curr.prev
#             curr.prev=curr.next
#             curr.next=temp
            curr= curr.next
#         print(curr.prev)
#         print(curr.next)
        self.head=curr
        
            
    def print(self):
        temp= self.head
        while(temp):
            print(temp.data, end=' ')
            temp=temp.next
```
